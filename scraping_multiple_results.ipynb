```python
from bs4 import BeautifulSoup
import requests
import json
import csv
import pandas as pd
```

```python
#Select the main url 
main_url="https://en.wikipedia.org/w/index.php?title=Category:Male_actors_in_Malayalam_cinema"
main_page=requests.get(main_url)
soup=BeautifulSoup(main_page.text,'html.parser')
category=soup.find('div', class_='mw-category')
list_of_names=category.text[1:].split('\n')

list_of_names
```

Procedure:

1. Since certain names differ with the respective homepages, those names are manually figured out and created as list of names called stopwords.

2. In the result1 variable only the infoboxes of "list of names" which is not present in the stopwords are extracted.

3. Next the names in stopwords variable is modified and created as a new variable called modified_names. 

4. IN the result2, the biography Vcard for the modified names is scraped.

5. The results are then stored in json format.

```python
stopwords = ['K. P. A. C. AzeezB','Dharmajan BolgattyC','Balachandran ChullikkaduD','DivyadarshanF','Vinay ForrtG',
            'Shivaji GuruvayoorH','Peter HeinI','Puneet IssarJ','Joemon JoshyK','Bobby Kottarakkara']
for word in list_of_names:
    if word in stopwords:
        list_of_names.remove(word)
```

```python
list_of_names
```

```python
result1=[]
for item in list_of_names:
    pg=wptools.page(item).get_parse()
    box_1=pg.data['infobox']
    result1.append(box_1)
with open('result_in_JSON_1.json', 'w') as f:
     json.dump(result1,f,indent=4)

    
    
```
